## Summary

This baseball dataset is provided by Udacity, it has total 1157 records with two dimensions called name and handedness, four measures including batting average, home runs, height and weight. To analyze the dataset, I plotted a pie chart to show the handedness distribution in all players, which shows that right handedness is dominant and takes 63.7% of all players. Then a boxplot and bar chart were both generated to show the batting average and home runs of all players by handedness. Using filters in horizontal histogram, I was able to find the bad and good batting and home runs by each player. Lastly, height and weight were plotted to see their relationship and their impact on batting average and home runs, and the results show that height is correlated to weight, and height is moderately related to batting average, but not to home runs.

## Design

I renamed the column of 'avg' to 'batting average' and 'HR' to 'home runs'. There are two dimensions and four measures in the baseball data. I first started to draw a histgram to see the percentage of each handedness in all players. I also plotted histagrams to see the big picture of the median batting average and median home runs among different handedness. Color mark was used to differentiate the three types of handedness.

A boxplot was created to see the performance of all the players within their handedness group. Many outliers were seen in the boxplot, so the median value of batting average and home runs would be used for the following figures. The boxplot was combined with handedness percentage histgram to form a dashbooard.

I created a set called low batting. Low batting was defined as any battering average less than the value of 0.23. I also created a hierarchy called handedness.name which was generated by combining the two columns of handedness and name. Based on these new columns, I plotted a horizontal histgram that moving battering average to column and handedness.name to column and using the handedness as filter and the low batting set to color mark. I then created another set called low home runs, which was defined as any home runs below 25. Then another horizontal histgram was created by setting home runs as colume and handedness.name as row and adding the handedness as filetr and loew homeruns as color mark.

A scatterplot was created to see the correlation of the height and weight. Handedness was used as filter and color mark. Name was used as detail mark.

A dashboard was created by combining the low batting and low home runs horizontal histgrams.

With some feedbacks, I added a pie chart to the story to show the handedness distribution. The pie chart size was adjusted to fit in the middle and a title was added. This plot was added in the beginning of the story to draw attention. A trend line was added to the scatterplot for the correlation of height and weight. Line charts was plotted to show the impact of height or weight on median batting average and median home runs by handedness. Trend lines was also added on the line charts. Lastly, the relationship between batting average and home runs were also plotted, handedness was used as filter and color and shape marks. A trend line was also added to show the r value.

According to the feedback from Udacity reviewer, I replaced L, B, R with Left, Right and Both. I also changed to a single color in bar charts. The last scatter plot was added with more trend lines.
